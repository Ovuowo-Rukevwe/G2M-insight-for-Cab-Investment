```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
Cab_Data = pd.read_csv("/content/Cab_Data.csv")
City = pd.read_csv("/content/City.csv")
Customer_ID = pd.read_csv("/content/Customer_ID.csv")
Transaction_ID = pd.read_csv("/content/Transaction_ID.csv")
```

```python
df_1 = pd.merge(Cab_Data, City , on="City")
```

```python
df_2 = pd.merge(Customer_ID, Transaction_ID, on="Customer ID")
```

```python
df_3 = pd.merge(df_1, df_2, on="Transaction ID")
df_3.head(3)
```

```python
def current_time(currents):
   dl = []
   for current in currents:
     try:
       x = datetime.date(1899,12,30)+datetime.timedelta(days=float(current))
       dl.append(x.strftime('%Y-%m-%d'))
     except:
        dl.append('n/a')
   return dl
```

```python
df_3['Date of Travel'] = current_time(df_3['Date of Travel'])
df_3['Date of Travel'] = pd.to_datetime(df_3['Date of Travel'])
```

```python
df_3.head(5)
```

```python
df_3['Year'] = df_3['Date of Travel'].dt.year
```

```python
df_3.head()
```

```python
# Getting the necessary info from the dataset
def data_info(data):
    df = pd.DataFrame(index=data.columns)
    df['Data_type'] = data.dtypes
    df['Total Value'] = data.count()
    df['Null_count'] = data.isnull().sum()
    df_1['Unique_count'] = data.nunique()
    return df
```

```python
data_info(df_3)
```

```python
df_3['Profits'] = df_3['Price Charged'] - df_3['Cost of Trip']

df_3['Profit/KM'] = df_3['Profits'] / df_3['KM Travelled']

df_3['Profit Margin'] = df_3['Cost of Trip'] / df_3['Price Charged']

df_3['Profit in percentage'] = ((df_3['Price Charged'] - df_3['Cost of Trip']) / df_3['Price Charged']*100)
```

```python
df_3.head()
```

```python
Yearly_profit = df_3.groupby(['Year','Company']).sum()[['Profits','Cost of Trip','Price Charged']].reset_index()
Yearly_profit
```

```python
Total_yearly_profit= Yearly_profit.groupby('Year').sum().reset_index()
Total_yearly_profit
```

```python
Yearly_profit_chart = Yearly_profit.sort_values(by=['Company'])
Yearly_profit_chart
```

```python
plot = sns.catplot(x="Year", y="Profits", hue="Company", col="Company",
                data=Yearly_profit_chart , kind="bar",palette=sns.color_palette(['pink', 'yellow'])
                );
plot.fig.subplots_adjust(top=0.8)
plot.fig.suptitle('Yealy profit in percentage')
plt.show()
```

```python
plot = sns.lineplot(data=Yearly_profit_chart, x="Year", y="Profits", hue="Company")
plot.set(title='Yearly profit')
plot.set(xticks=([2016, 2017, 2018]))
plt.show() 
```

```python
Average_profit_in_KM = df_3.groupby(['Company','Year']).mean()['Profit/KM'].unstack('Company').reset_index()
Average_profit_in_KM
```

```python
Average_profit_in_KM.plot(x='Year',kind='bar',stacked=True,title='Average Profit per KM',color=['pink','yellow'])
plt.show()
```

```python
Profit_in_company = df_3.groupby('Company').sum()['Profits'].reset_index()

Number_of_Rides = df_3.groupby('Company').count()['Transaction ID'].reset_index()

Table_merge_on_profit_and_rides = pd.merge(Profit_in_company, Number_of_Rides, on='Company')
Table_merge_on_profit_and_rides

# Transaction ID is used for the Total number of rides
```

```python
Table_merge_on_profit_and_rides.rename(columns={"Transaction ID": "Total_number_of_Rides"},inplace=True)

Table_merge_on_profit_and_rides['Profit/Ride'] = Table_merge_on_profit_and_rides['Profits']/Table_merge_on_profit_and_rides['Total_number_of_Rides']
Table_merge_on_profit_and_rides
```

```python
fig,axes = plt.subplots(1,3)
fig.set_size_inches(20,20)
axes[0].pie(Table_merge_on_profit_and_rides['Total_number_of_Rides'],labels=Table_merge_on_profit_and_rides['Company'],autopct='%1.1f%%',shadow=True,startangle=45,)
axes[0].set_title('Total Rides in Percentage')
axes[1].pie(Table_merge_on_profit_and_rides['Profits'],labels=Table_merge_on_profit_and_rides['Company'],autopct='%1.1f%%',shadow=True,startangle=90)
axes[1].set_title('Total Rides in Percentage')
axes[2].pie(Table_merge_on_profit_and_rides['Profit/Ride'],labels=Table_merge_on_profit_and_rides['Company'],autopct='%1.1f%%',shadow=True,startangle=180)
axes[2].set_title('Total Rides in Percentage')
```
